Script:

```raku
unit sub MAIN(Str:D $s);

use MONKEY-SEE-NO-EVAL;

my (@matrix, $k);
EVAL $s;

with @matrix[0] ~~ Array ?? +@matrix[0] !! +@matrix -> \cols {
    put @matrix[*;*].rotate(-$k).rotor(cols).map(*.Array.raku).join(', '); 
}
```

`@matrix[*;*]` flattens the matrix.
`rotate($i)` shifts an array left by `$i` elements. We need to shift the array right, so a negative number is provied for the offset.
Then the array is **unflattend**, and printed.
`cols` is the number of columns of the matrix.
