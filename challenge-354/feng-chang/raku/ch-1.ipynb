Script:

```raku
unit sub MAIN(*@ints);

my @pairs = +«@ints.sort.rotor(2 => -1).map(*.Array);
my $min-diff = @pairs.map({ -([-] @$_) }).min;
put @pairs.grep({ -([-] @$_) == $min-diff }).map({ "[{.join(', ')}]" }).join(', ');
```

`+«@ints` converts all @ints elements to integer. Then the array is sorted, and we store all candidate pairs into `@pairs` array.

```raku
my @ints = <4 2 1 3>;
put @ints.raku;
put (+«@ints).raku;
put (+«@ints).sort.raku;
put (+«@ints).sort.rotor(2 => -1).raku;
put (+«@ints).sort.rotor(2 => -1).map(*.Array).raku;
```

For each pair, `@$_` is the pair array. `[-] @$_` is the same as `@$_[0] - @$_[1]`.
The minimum of the difference is stored in `$min-diff`.

In the last statement, we `grep` all closest pairs and print them.
