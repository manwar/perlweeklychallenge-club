###### PWC365, Task 2 Valid Token Counter
**冯昶 2026-03-21**

Script:

```raku
unit sub MAIN(*@tokens);

put @tokens.map({ +?.match(/^ <[a..z]>+ [ '-' <[a..z]>+ ]? <:P - [-]>? $/) }).sum;
```

First, `:P` stands for `<punct>`, and `<:P - [-]>` is the character set that is punctuations minus '-'.

Second, to count the matches, you have to cast a match result into boolean value with '?', then cast the boolean into integer with '+'.
