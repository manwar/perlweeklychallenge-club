.comb translates a string into an array of characters, and .Bag translates the array into hash, in which direction is the key, and counts of the direction is value:

```raku
put 'ULDRULD'.comb.Bag.raku;
```

If and only if the counts of U(up) equals D(down) and L(left) equals R(right), it will come back to the start point.

Here is my Raku script for task 2:

```raku
unit sub MAIN(Str:D $s);

put [.<U>,.<L>] eqv [.<D>,.<R>] with $s.comb.Bag;
```
