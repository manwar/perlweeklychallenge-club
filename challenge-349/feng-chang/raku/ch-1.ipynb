Regex `/./` matches the first occurance of a single character:

```raku
put 'abc'.match(/./).Str;
```

Regex `/(.)$0*/` matches the first occurance of a string consisting of a unique letter:

```raku
put 'aaabc'.match(/(.)$0*/).Str;
```

In which `(.)` captures a character, and `$0` refers to the captured character.

If we add an `:ex` proverb to the regex, it will match all possible candidates:

```raku
put 'aaabcc'.match(/:r (.)$0*/, :ex)».Str;
```

Here is my ch-1.raku script:

```raku
unit sub MAIN(Str:D $s);

put $s.match(/(.)$0*/, :ex)».chars.max;
```
