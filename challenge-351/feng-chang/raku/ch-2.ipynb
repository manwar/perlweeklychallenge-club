Script:

```raku
unit sub MAIN(*@nums);

@nums = +«@nums.sort;
put so (@nums Z (|@nums[^2] ... *)[^+@nums]).map({ [==] $_ }).all;
```

`+«@nums` converts each element of @nums array from string into integer.

```raku
my @a = 1,2 ... *;      # Raku can figure out an entire arithmetic progression as long as you give the first 2 elements
put @a[^10];            # print the first 10 elements of the arithmetic progression

my @b = 1,2.5 ... *;    # another example with rationals
put @b[^7];
```

We first generate an arithmetic progression by giving the first 2 elements by `|@nums[^2] ... *`.
Then we compare the two sequences one element by one element, if all the pairs are equal, then the two seqquences are identical.
And it means the sorted `@nums` array is an arithmetic progression.

```raku
put (<a b c> Z <X Y Z>).raku;       # the Z operator zips together two arrays, and output an array of pairs.
put [==] 1, 2;                      # False
put [==] 2.5, 2.5;                  # True
put [==] 6, 6, 6;                   # True
```
