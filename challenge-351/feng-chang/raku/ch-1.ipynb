My script for task 1:

```raku
#!/bin/env raku

unit sub MAIN(*@ints);

my ($min, $max) = @ints.min, @ints.max;
@ints .= grep(* != ($min,$max).any);
put +@ints > 0 ?? @ints.sum/+@ints !! 0;
```

First, we find the minimum and maximum values of our array.

Then we get rid of all min/max values with `@ints .= grep(* != ($min,$max).any);`
It is the same as `@ints = @ints.grep(* != ($min,$max).any);`

Finally, we print the average of the array. When the array is empty, we simply print 0 to avoid the divide-by-zero error.
