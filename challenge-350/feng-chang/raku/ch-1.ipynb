.comb() translates a string into a list of characters.
.rotor(3) consumes 3 elements of the list each time.
.rotor(3 => -2) consumes 3 elements of the list, and start the next iteration by going back 2 elements each time.
.Bag() translates a list into an hash, in which elements are keys, the counts of each element are values.
If a list has N distinct elements, .Bag() will have N pairs.

```raku
put 'abcaefg'.comb.gist;                  # (a b c a e f g)
put <a b c a e f g h i>.rotor(3).gist;    # ((a b c) (a e f) (g h i))
put <a b c a e f g>.rotor(3 => -2).gist;  # ((a b c) (b c a) (c a e) (a e f) (e f g))
put <a a a>.Bag.raku;   # ('a' => 3)
put <a b a>.Bag.raku;   # ('a' => 2, 'b' => 1)
put <a b c>.Bag.raku;   # ('a' => 1, 'b' => 1, 'c' => 1)
```

Combine all the above together, here is my scprit for Task 1:

```raku
unit sub MAIN(Str:D $s);

# print the counts of all 3 character substrings that has 3 distinct letters.
put +$s.comb.rotor(3 => -2).grep(*.Bag.elems == 3);
```
