To test if 2 different numbers have the same digits, we can reorder all the digits of each number from low to high, and test if the results are equal.
Note we test equalness with 'eq' instead of '==' to avoid pit falls.

```raku
put '13254'.comb.sort.join;                # 12345
put '13254'.comb.sort.join eq '12345';     # True

put 15*7;                                  # 15 * 7 = 105
put (15*7).comb.sort.join;                 # 015
put (15*7).comb.sort.join eq '15';         # False, '015' and '15' are different
put (15*7).comb.sort.join == 15;           # True, 015 equals 15 as numbers
```

Here is my script for Task 2:

```raku
unit sub MAIN(UInt:D $from, UInt:D $to, UInt:D $count);

put +($from..$to).grep(-> $n {
    my $b = $n.comb.sort.join;
    +(2..9).grep({ ($n*$_).comb.sort.join eq $b }) ≥ $count
});
```

We test 2 through 9 times of each candidate `$n`, whether the result is a shuffle pair of `$n`.
Each number with no less than $count shuffle pairs is recorded.
Finally, we print the counts of numbers that were recorded.
