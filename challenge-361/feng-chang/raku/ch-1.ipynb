Script:

```raku
unit sub MAIN(UInt:D $int is copy where * ≤ 100);

my @fib = 1, 2, * + * ...^ * > 100;

my @answer;
my $ndx = @fib.first(* ≤ $int, :k, :end);

loop {
    @answer.push(@fib[$ndx]);
    $int -= @answer.tail;
    last if $int ≤ 0;
    $ndx = @fib[^($ndx-1)].first(* ≤ $int, :k, :end);
}

put @answer.join(',');
```

2 things to note. First, the usage of `...^`:

```raku
my @fib1 = 1, 2, * + * ... *;          # generates a lazy fibonacci sery
my @fib2 = 1, 2, * + * ... * > 100;    # generates an array in which the last number is the first fabonacci number that is > 100
put @fib2.gist;
my @fib3 = 1, 2, * + * ...^ * > 100;   # generates an array that contains the numbers that is ≤ 100
put @fib3.gist;
```

Second, the usage of `first()`. `* ≤ $int` is the search condition. The `:k` proverb tells `first()` to return the key of the first number instead of its value. The `:end` proverb makes the search start from the end to the start, instead of from start to end.
Example:

```raku
my @a = 10,20,30,40,30,50;
put @a.first(30);               # print the found number
put @a.first(30, :k);           # the index of first 30 from left to right is 2
put @a.first(30, :k, :end);     # the index of first 30 from right to left is 4
```
