Script:

```raku
unit sub MAIN(Int:D $int);

put $int.flip.comb.rotor(3, :partial)».join.join(',').flip;
```

```raku
put 12345.flip;                                                # first, flip the number string
put 12345.flip.comb.raku;                                      # array of all the digits
put 12345.flip.comb.rotor(3).raku;                             # group all 3 consecutive digits
put 12345.flip.comb.rotor(3, :partial).raku;                   # we need all the digits, including the leading 1 or 2 digits
put 12345.flip.comb.rotor(3, :partial)».join.raku;             # join every group
put 12345.flip.comb.rotor(3, :partial)».join.join(',');        # join all the groups with ','
put 12345.flip.comb.rotor(3, :partial)».join.join(',').flip;   # flip back
```
