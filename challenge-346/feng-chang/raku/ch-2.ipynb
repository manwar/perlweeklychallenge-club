Following is my ch-2.raku script:

```raku
#!/bin/env raku

unit sub MAIN(Str:D $s, Int:D $tgt);

use MONKEY-SEE-NO-EVAL;

my @s = $s.comb;
my @opers = '+', '-', '*', '';

my @ans;
for [X] @opers xx (+@s-1) -> @ops {
    my $exp = roundrobin(@s, @ops).flat.join;
    next if $exp ~~ /« '0' \d/;    # drop expressions which contains number with a leading 0
    @ans.push($exp) if EVAL($exp) == $tgt;
}
put @ans.join(', ');
```

Let me explain the code line by line.

```raku
# $s.comb translates string $s into list of numbers
my $s = '1234';
my @s = $s.comb;
put @s.raku;
```

```raku
# @opers array consists of all possible operations
my @opers = '+', '-', '*', '';
# of which the '' op is the operation to join 2 numbers together
```

```raku
# +@s is the size of array @s, while +@s is size of @s minus 1
my @s = <1 2 3 4 5>;
put +@s-1;
```

`[X] @opers xx 3` is the same as `@opers X @opers X @opers`.
So @ops in the for loop iterates through all possible operation combinations.

roundrobin() combines the digits and operations together. e.g.,

```raku
put roundrobin(<1 2 3 4 5>, <+ - * +>).flat.join;
```

```raku
# the « sign in regex is left anchor for a word
next if $exp ~~ /« '0' \d/;    # drop expressions which contains number with a leading 0
```

```raku
# at the end of the loop, @ans will contain all answers
@ans.push($exp) if EVAL($exp) == $tgt;

`use MONKEY-SEE-NO-EVAL;` must be declared because EVAL() is considered unsafe.

The last line prints all answers separated by commas:
```

```raku
put @ans.join(', ');
```
