First, let's define a method to translate hh:mm format time into minutes:

```raku
my method time-to-min(Str:D $time : --> UInt:D) {
    .substr(0,2) * 60 + .substr(3) with $time
}
put '00:01'.&time-to-min;
put '01:00'.&time-to-min;
```

```raku
Then, an introduction to the `polymod()` function.

polymod() is effective in translating numbers into units, e.g.:
```

```raku
# translate 99999 seconds into number of seconds, minutes, hours, and days
put 99999.polymod(60,60,24).raku;       # 1 day, 3 hours, 46 minutes, and 39 seconds

# translate 12345 into digits:
put 12345.polymod(10 xx 4).raku;
```

Here is my raku script for task 2:

```raku
unit sub MAIN(Str:D $source, Str:D $target);

my method time-to-min(Str:D $time : --> UInt:D) {
    .substr(0,2) * 60 + .substr(3) with $time
}

my $diff = $target.&time-to-min - $source.&time-to-min;
$diff += 60 * 24 if $diff < 0;       # add a day to the time difference when target is from the 2nd day
put $diff.polymod(5, 3, 4).sum;      # first unit is 5min, 2nd unit is 5*3=15min, 3rd unit is 15*4=60min
```
