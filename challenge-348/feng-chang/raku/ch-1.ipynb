@vowels is an array of all the vowels:

```raku
my @vowels = <a e i o u>;
put @vowels.raku;
```

.lc translate all capital letters of a string into lower case letters:

```raku
my $w = 'TeXtBoOk'.lc;
put $w;
```

.comb translate a string into an array of characters:

```raku
put 'textbook'.comb.raku;
```

`+@w` is the same as `@w.elems`, and +@w/2 is half of the size of the array @w:

```raku
my @w = 'textbook'.comb;
put +@w/2;
```

`@w[^+@w/2]` is first half of @w, while `@w[+@w/2..*]` is the second half.

`.grep(@vowels.any)` find all vowles in the array:

```raku
my @vowels = <a e i o u>;
my @w = 'textbook'.comb;
put @w.grep(@vowels.any).raku;
put +@w.grep(@vowels.any);
```

Now, let's put all of the above together:

```raku
my @vowels = <a e i o u>;
my @w = 'TeXtBoOk'.lc.comb;
put +@w[^+@w/2].grep(@vowels.any) == +@w[+@w/2..*].grep(@vowels.any);
```
