# PWC352, Task 1 Match String 
**冯昶 2025-12-15**

Script:

```raku
unit sub MAIN(*@str);

@str .= unique;
put @str.grep(-> $s { +@str.grep(*.contains($s)) > 1 });
```

Explanation:

`@str .= unique` is the same as `@str = @str.unique`.

Since every string contains itself, so we print all strings that are **contained** by at least 2 strings in the array, that happens to be the answer.
