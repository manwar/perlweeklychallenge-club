## PWC352, Task 2 Binary Prefix 
**冯昶 2025-12-15**

Script:

```raku
unit sub MAIN(*@bin);

put ([\~] @bin).map((*.parse-base(2) %% 5).lc).join(', ');
```

Explanation:

```raku
put ([~] <a b c d>).raku;   # `[~] @array` is the same as @array.join
```

```raku
put ([\~] <a b c d>).raku;  # put a slash before the operator and we get all the intermediate results.
```

```raku
put '101'.parse-base(2);             # 5
put '101'.parse-base(2) %% 5;        # True
put ('101'.parse-base(2) %% 5).lc;   # 'true'
```
