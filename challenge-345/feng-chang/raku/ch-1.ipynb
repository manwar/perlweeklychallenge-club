(1..+@ints-2) is the range of indices from the 2nd element to the 2nd to last element.

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).raku;
```

The candidates for peak positions:

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).map({ @ints[$_] }).raku;
```

`(@ints[$_-1], @ints[$_+1])` are the two neighbours of each element:

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).map({ (@ints[$_-1], @ints[$_+1]).all }).raku;
```

Now, we got an array of junctions of comparison results:

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).map({ @ints[$_] > (@ints[$_-1], @ints[$_+1]).all }).raku;
```

Let's put a '**so**' before each junction to normalize it to a bool value, and see the results more clearly:

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).map({ so @ints[$_] > (@ints[$_-1], @ints[$_+1]).all }).raku;
```

Change **map** to **grep** and get the final answer:

```raku
my @ints = 1,2,3,2,4,1;
put (1..+@ints-2).grep({ @ints[$_] > (@ints[$_-1], @ints[$_+1]).all }).raku;
```

Cover the case for zero result:

```raku
my @ints = 1,2,3,2,4,1;
put do { +$_ ?? $_ !! 0 } with (1..+@ints-2).grep({ @ints[$_] > (@ints[$_-1], @ints[$_+1]).all });
```
